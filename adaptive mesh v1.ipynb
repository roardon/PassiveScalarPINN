```python
import numpy as np
import deepxde as dde
import matplotlib.pyplot as plt
from copy import deepcopy
import scipy.io
import sys
import math
import os
import scipy.interpolate

reference_data = np.loadtxt("cylinder_point_source.dat")
x_cell_centroid = reference_data[:, 0]
y_cell_centroid = reference_data[:, 1]
vol_weights = reference_data[:, 3]
u_mean_x = reference_data[:, 4]
u_mean_y = reference_data[:, 5]
t_mean = reference_data[:, 6]
div_uprime2_mean_x = reference_data[:, 10]
div_uprime2_mean_y = reference_data[:, 11]
g_mean = reference_data[:, 14]
Re = 150
Pr = 266.8

iters = 5
adamiters = 4000

def pde(X, V):
    # put quantities into arrays of single-value arrays
    u = V[:, 0:1]
    v = V[:, 1:2]
    p = V[:, 2:3]
    t = V[:, 3:4]
    fx = V[:, 4:5]
    fy = V[:, 5:6]
    g = V[:, 6:7]
    
    # compute partial derivatives. components i, j: i is the component from the V matrix (0 - 6 for respective properties),
    # j is component for spatial coordinate (0 - 1 for x or y)
    du_x = dde.grad.jacobian(V, X, i=0, j=0)
    dv_y = dde.grad.jacobian(V, X, i=1, j=1)
    du_y = dde.grad.jacobian(V, X, i=0, j=1)
    dv_x = dde.grad.jacobian(V, X, i=1, j=0)
    dp_x = dde.grad.jacobian(V, X, i=2, j=0)
    dp_y = dde.grad.jacobian(V, X, i=2, j=1)
    dt_x = dde.grad.jacobian(V, X, i=3, j=0)
    dt_y = dde.grad.jacobian(V, X, i=3, j=1)
    # compute partial second derivatives. First parameter is variable rather than matrix.
    # i, j are denominator spatial coordinate indices, e.g dy/dxidxj
    du_xx = dde.grad.hessian(u, X, i=0, j=0)
    dv_xx = dde.grad.hessian(v, X, i=0, j=0)
    du_yy = dde.grad.hessian(u, X, i=1, j=1)
    dv_yy = dde.grad.hessian(v, X, i=1, j=1)
    dt_xx = dde.grad.hessian(t, X, i=0, j=0)
    dt_yy = dde.grad.hessian(t, X, i=1, j=1)
    # compute partial derivatives of forcing term
    dfx_x = dde.grad.jacobian(V, X, i=4, j=0)
    dfy_y = dde.grad.jacobian(V, X, i=5, j=1)
    
    return [
        du_x + dv_y,
        u * du_x + v * du_y + dp_x - (1.0 / Re) * (du_xx + du_yy) + fx,
        u * dv_x + v * dv_y + dp_y - (1.0 / Re) * (dv_xx + dv_yy) + fy,
        dfx_x + dfy_y, 
        u * dt_x + v * dt_y - (1.0 / (Re * Pr)) * (dt_xx + dt_yy) + g
    ]

vertex_botleft = [-3, -4]
vertex_topright = [15, 4]
geom = dde.geometry.Rectangle(vertex_botleft, vertex_topright) - dde.geometry.Disk([0, 0], 0.5)
figsize = (16, 12)

Nx = int((vertex_topright[0]-vertex_botleft[0])*100)+1
Ny = int((vertex_topright[1]-vertex_botleft[1])*100)+1

def generate_training_points(x, y, u, v, t):
    x_t = []
    y_t = []
    u_t = []
    v_t = []
    t_t = []
    
    x_t = np.array(x_t)
    y_t = np.array(y_t)
    u_t = np.array(u_t)
    v_t = np.array(v_t)
    t_t = np.array(t_t)
    
    x_t = x[0::1].reshape(-1, 1)
    y_t = y[0::1].reshape(-1, 1)
    u_t = u[0::1].reshape(-1, 1)
    v_t = v[0::1].reshape(-1, 1)
    t_t = t[0::1].reshape(-1, 1)
    
    X = []
    
    for i in range(x_t.shape[0]):
        if geom.inside([x_t[i, 0], y_t[i, 0]]) \
                and x_t[i, 0] > vertex_botleft[0] and x_t[i, 0] < vertex_topright[0] \
                and y_t[i, 0] > vertex_botleft[1] and y_t[i, 0] < vertex_topright[1]:
            X.append([x_t[i, 0], y_t[i, 0], u_t[i, 0], v_t[i, 0], t_t[i, 0]])
    
    X = np.array(X)
 
    return np.hsplit(X, 5)

    
def func_zeros(X):
    x = X[:, 0:1]
    return x * 0

def boundary(x, on_boundary):
    return on_boundary and not(
        np.isclose(x[0], vertex_botleft[0]) or np.isclose(x[0], vertex_topright[0]) 
        or np.isclose(x[1], vertex_botleft[1]) or np.isclose(x[1], vertex_topright[1])
    )
```

```python
points = np.zeros((len(x_cell_centroid), 2))
for i in range(len(x_cell_centroid)):
    points[i][0] = x_cell_centroid[i]
    points[i][1] = y_cell_centroid[i]

# integer before 'j' = (number of points in x or y) + 1    
    
grid_x, grid_y = np.mgrid[-14.9167:39.9:44j, -9.95:9.95:16j]
num_x = np.shape(grid_x)[0]
num_y = np.shape(grid_y)[1]

gridu = scipy.interpolate.griddata(points, u_mean_x, (grid_x, grid_y)).T
gridv = scipy.interpolate.griddata(points, u_mean_y, (grid_x, grid_y)).T

# now need to generate new x and y points using linspace

x_line = np.linspace(-14.9167, 39.9, num_x)
y_line = np.linspace(-9.95, 9.95, num_y)
    
# reformat new_grid u data to be same shape as x and y

current_y = np.zeros(num_x)
new_x = []
new_y = []
new_u = []
new_v = []
for i in range(num_y):
    for j in range(num_x):
        current_y[j] = y_line[i]
    new_y = np.concatenate((new_y, current_y))
    new_x = np.concatenate((new_x, x_line))
    new_u = np.concatenate((new_u, gridu[i]))
    new_v = np.concatenate((new_v, gridv[i]))
    
points = np.zeros((len(x_cell_centroid), 2))
for i in range(len(x_cell_centroid)):
    points[i][0] = x_cell_centroid[i]
    points[i][1] = y_cell_centroid[i]

# integer before 'j' = (number of points in x or y) + 1    
    
grid_x2, grid_y2 = np.mgrid[-14.9167:39.9:660j, -9.95:9.95:240j]
num_x2 = np.shape(grid_x2)[0]
num_y2 = np.shape(grid_y2)[1]

gridt = scipy.interpolate.griddata(points, t_mean, (grid_x2, grid_y2)).T

# now need to generate new x and y points using linspace

x_line2 = np.linspace(-14.9167, 39.9, num_x2)
y_line2 = np.linspace(-9.95, 9.95, num_y2)
    
# reformat new_grid u data to be same shape as x and y

current_y2 = np.zeros(num_x2)
new_x2 = []
new_y2 = []
new_t = []
for i in range(num_y2):
    for j in range(num_x2):
        current_y2[j] = y_line2[i]
    new_y2 = np.concatenate((new_y2, current_y2))
    new_x2 = np.concatenate((new_x2, x_line2))
    new_t = np.concatenate((new_t, gridt[i]))
    

new_x = np.array(new_x)
new_y = np.array(new_y)
new_x2 = np.array(new_x2)
new_y2 = np.array(new_y2)
new_u = np.array(new_u)
new_v = np.array(new_v)
new_t = np.array(new_t)

[vel_x_training, vel_y_training, u_training, v_training, dummy_t] = \
    generate_training_points(new_x, new_y, new_u, new_v, new_u)

vel_training_points = np.hstack((vel_x_training, vel_y_training))

[t_x_training, t_y_training, dummy_u, dummy_v, t_training] = \
    generate_training_points(new_x2, new_y2, new_t, new_t, new_t)

temp_training_points = np.hstack((t_x_training, t_y_training))

training_points = np.concatenate((temp_training_points, vel_training_points))

txu = []
tyu = []
txt = []
tyt = []
for i in range(0, len(vel_training_points)):
    txu.append(vel_training_points[i][0])
    tyu.append(vel_training_points[i][1])
for i in range(0, len(temp_training_points)):
    txt.append(temp_training_points[i][0])
    tyt.append(temp_training_points[i][1])
    
cylinderx = np.linspace(-0.5, 0.5, 50)
cylindery = np.sqrt(0.5**2 - cylinderx**2)
plt.figure(figsize=figsize)
plt.scatter(txu, tyu, s=1.4, label="Velocity data points")
plt.scatter(txt, tyt, s=1.4, label="Scalar data points")
plt.plot(cylinderx, cylindery, color='k')
plt.plot(cylinderx, -cylindery, color='k')
axes = plt.gca()
axes.set_aspect(1)
axes.set_xlim([-3.0, 15.0])
axes.set_ylim([-4.0, 4.0])
axes.set_title("Data point distribution")
plt.xlabel("x/c")
plt.ylabel("y/c")
plt.legend(loc=(1.05, 0.45))
plt.show()
plt.close()
print(str(len(training_points)) + " training points in domain")

# u, v dirichlet BCs at each point in training_points
u_training_points = dde.PointSetBC(vel_training_points, u_training, component=0)
v_training_points = dde.PointSetBC(vel_training_points, v_training, component=1)
t_training_points = dde.PointSetBC(temp_training_points, t_training, component=3)
# dirichlet BCs at walls, component indices from pde function
bc_wall_u = dde.DirichletBC(geom, func_zeros, boundary, component=0)
bc_wall_v = dde.DirichletBC(geom, func_zeros, boundary, component=1)
bc_wall_t = dde.NeumannBC(geom, func_zeros, boundary, component=3)
bc_wall_fx = dde.DirichletBC(geom, func_zeros, boundary, component=4)
bc_wall_fy = dde.DirichletBC(geom, func_zeros, boundary, component=5)
bc_wall_g = dde.NeumannBC(geom, func_zeros, boundary, component=6)

data = dde.data.PDE(
    geom, pde, 
    [bc_wall_u, bc_wall_v, bc_wall_t, bc_wall_fx, bc_wall_fy, bc_wall_g, u_training_points, v_training_points, t_training_points], 
    35000, 2500, solution=None, num_test=10, 
)

layer_size = [2] + [100] * 7 + [7]
activation = "tanh"
initializer = "Glorot uniform"
net = dde.maps.FNN(layer_size, activation, initializer)

model = dde.Model(data, net)

loss_weights = [1, 1, 1, 1, 1, 10, 10, 10, 10, 10, 10, 10, 10, 10]
model.compile("adam", lr=0.001, loss_weights=loss_weights)

for k in range(iters):

    x_plot = np.linspace(vertex_botleft[0], vertex_topright[0], Nx)
    y_plot = np.linspace(vertex_botleft[1], vertex_topright[1], Ny)

    # define prediction points
    z = np.array([np.array([i, j]) for i in x_plot for j in y_plot])

    # predict fields
    y = model.predict(z)
    u_star = y[:, 0][:, None]
    v_star = y[:, 1][:, None]
    p_star = y[:, 2][:, None]
    t_star = y[:, 3][:, None]
    fx_star = y[:, 4][:, None]
    fy_star = y[:, 5][:, None]
    g_star = y[:, 6][:, None]

    u_star_data = deepcopy(u_star)
    v_star_data = deepcopy(v_star)
    p_star_data = deepcopy(p_star)
    t_star_data = deepcopy(t_star)
    fx_star_data = deepcopy(fx_star)
    fy_star_data = deepcopy(fy_star)
    g_star_data = deepcopy(g_star)

    # remove inside points
    bad_indices = []
    for i in range(len(z)):
        if(np.sqrt(z[i][0]**2 + z[i][1]**2) < 0.5):
            bad_indices.append(i)

    for i in range(len(bad_indices)):
        u_star_data[bad_indices[i]][0] = 0
        v_star_data[bad_indices[i]][0] = 0
        p_star_data[bad_indices[i]][0] = 0
        t_star_data[bad_indices[i]][0] = 0
        fx_star_data[bad_indices[i]][0] = 0
        fy_star_data[bad_indices[i]][0] = 0
        g_star_data[bad_indices[i]][0] = 0

    u_star_data = u_star_data.reshape(Nx, Ny).T
    v_star_data = v_star_data.reshape(Nx, Ny).T
    p_star_data = p_star_data.reshape(Nx, Ny).T
    t_star_data = t_star_data.reshape(Nx, Ny).T
    fx_star_data = fx_star_data.reshape(Nx, Ny).T
    fy_star_data = fy_star_data.reshape(Nx, Ny).T
    g_star_data = g_star_data.reshape(Nx, Ny).T

    X, Y = np.meshgrid(x_plot, y_plot)

    # get points of highest scalar concentration

    t_finding = deepcopy(t_star_data)
    u_grad = np.array(np.gradient(u_star_data, 0.01))
    v_grad = np.array(np.gradient(v_star_data, 0.01))

    u_grad_x = u_grad[1, :, :]
    v_grad_x = v_grad[1, :, :]
    u_grad_y = u_grad[0, :, :]
    v_grad_y = v_grad[0, :, :]

    u_grad = np.log10(np.sqrt(u_grad_x**2 + u_grad_y**2))
    v_grad = np.log10(np.sqrt(v_grad_x**2 + v_grad_y**2))

    velocity_curl = np.log10(np.abs(v_grad_x - u_grad_y))

    max_ts = np.zeros(100)
    max_x1 = np.zeros(100)
    max_y1 = np.zeros(100)
    
    max_ugrad = np.zeros(100)
    max_x2 = np.zeros(100)
    max_y2 = np.zeros(100)
    
    max_vgrad = np.zeros(100)
    max_x3 = np.zeros(100)
    max_y3 = np.zeros(100)
    
    max_curl = np.zeros(100)
    max_x4 = np.zeros(100)
    max_y4 = np.zeros(100)

    for i in range(len(max_ts)):
        max_ts[i] = np.max(t_finding)
        max_x1[i] = X[np.unravel_index(np.argmax(t_finding), t_finding.shape)]
        max_y1[i] = Y[np.unravel_index(np.argmax(t_finding), t_finding.shape)]
        t_finding[np.unravel_index(np.argmax(t_finding), t_finding.shape)] = 0
        max_ugrad[i] = np.max(t_finding)
        max_x2[i] = X[np.unravel_index(np.argmax(u_grad), u_grad.shape)]
        max_y2[i] = Y[np.unravel_index(np.argmax(u_grad), u_grad.shape)]
        u_grad[np.unravel_index(np.argmax(u_grad), u_grad.shape)] = 0
        max_vgrad[i] = np.max(v_grad)
        max_x3[i] = X[np.unravel_index(np.argmax(v_grad), v_grad.shape)]
        max_y3[i] = Y[np.unravel_index(np.argmax(v_grad), v_grad.shape)]
        v_grad[np.unravel_index(np.argmax(v_grad), v_grad.shape)] = 0
        max_curl[i] = np.max(velocity_curl)
        max_x4[i] = X[np.unravel_index(np.argmax(velocity_curl), velocity_curl.shape)]
        max_y4[i] = Y[np.unravel_index(np.argmax(velocity_curl), velocity_curl.shape)]
        velocity_curl[np.unravel_index(np.argmax(velocity_curl), velocity_curl.shape)] = 0

    new_grid_centre1 = [np.mean(max_x1), np.mean(max_y1)]

    new_grid_x_range1 = [new_grid_centre1[0]-0.8*(1.4**(4-k)), new_grid_centre1[0]+0.8*(1.4**(4-k))]
    new_grid_y_range1 = [new_grid_centre1[1]-0.8*(1.4**(4-k)), new_grid_centre1[1]+0.8*(1.4**(4-k))]

    grid_x2, grid_y2 = np.mgrid[new_grid_x_range1[0]:new_grid_x_range1[1]:9j, new_grid_y_range1[0]:new_grid_y_range1[1]:9j]
    num_x2 = np.shape(grid_x2)[0]
    num_y2 = np.shape(grid_y2)[1]

    gridu2 = scipy.interpolate.griddata(points, u_mean_x, (grid_x2, grid_y2)).T
    gridv2 = scipy.interpolate.griddata(points, u_mean_y, (grid_x2, grid_y2)).T

    # now need to generate new x and y points using linspace

    x_line2 = np.linspace(new_grid_x_range1[0], new_grid_x_range1[1], num_x2)
    y_line2 = np.linspace(new_grid_y_range1[0], new_grid_y_range1[1], num_y2)

    # reformat new_grid u data to be same shape as x and y

    current_y2 = np.zeros(num_x2)
    new_x2 = []
    new_y2 = []
    new_u2 = []
    new_v2 = []
    for i in range(num_y2):
        for j in range(num_x2):
            current_y2[j] = y_line2[i]
        new_y2 = np.concatenate((new_y2, current_y2))
        new_x2 = np.concatenate((new_x2, x_line2))
        new_u2 = np.concatenate((new_u2, gridu2[i]))
        new_v2 = np.concatenate((new_v2, gridv2[i]))
        
    new_grid_centre2 = [np.mean(max_x2), np.mean(max_y2)]

    new_grid_x_range2 = [new_grid_centre2[0]-(1.4**(4-k)), new_grid_centre2[0]+(1.4**(4-k))]
    new_grid_y_range2 = [new_grid_centre2[1]-(1.4**(4-k)), new_grid_centre2[1]+(1.4**(4-k))]

    grid_x3, grid_y3 = np.mgrid[new_grid_x_range2[0]:new_grid_x_range2[1]:4j, new_grid_y_range2[0]:new_grid_y_range2[1]:4j]
    num_x3 = np.shape(grid_x3)[0]
    num_y3 = np.shape(grid_y3)[1]

    gridu3 = scipy.interpolate.griddata(points, u_mean_x, (grid_x3, grid_y3)).T
    gridv3 = scipy.interpolate.griddata(points, u_mean_y, (grid_x3, grid_y3)).T

    # now need to generate new x and y points using linspace

    x_line3 = np.linspace(new_grid_x_range2[0], new_grid_x_range2[1], num_x3)
    y_line3 = np.linspace(new_grid_y_range2[0], new_grid_y_range2[1], num_y3)

    # reformat new_grid u data to be same shape as x and y

    current_y3 = np.zeros(num_x3)
    new_x3 = []
    new_y3 = []
    new_u3 = []
    new_v3 = []
    for i in range(num_y3):
        for j in range(num_x3):
            current_y3[j] = y_line3[i]
        new_y3 = np.concatenate((new_y3, current_y3))
        new_x3 = np.concatenate((new_x3, x_line3))
        new_u3 = np.concatenate((new_u3, gridu3[i]))
        new_v3 = np.concatenate((new_v3, gridv3[i]))
        
    new_grid_centre3 = [np.mean(max_x3), np.mean(max_y3)]

    new_grid_x_range3 = [new_grid_centre3[0]-(1.4**(4-k)), new_grid_centre3[0]+(1.4**(4-k))]
    new_grid_y_range3 = [new_grid_centre3[1]-(1.4**(4-k)), new_grid_centre3[1]+(1.4**(4-k))]

    grid_x4, grid_y4 = np.mgrid[new_grid_x_range3[0]:new_grid_x_range3[1]:4j, new_grid_y_range3[0]:new_grid_y_range3[1]:4j]
    num_x4 = np.shape(grid_x4)[0]
    num_y4 = np.shape(grid_y4)[1]

    gridu4 = scipy.interpolate.griddata(points, u_mean_x, (grid_x4, grid_y4)).T
    gridv4 = scipy.interpolate.griddata(points, u_mean_y, (grid_x4, grid_y4)).T

    # now need to generate new x and y points using linspace

    x_line4 = np.linspace(new_grid_x_range3[0], new_grid_x_range3[1], num_x4)
    y_line4 = np.linspace(new_grid_y_range3[0], new_grid_y_range3[1], num_y4)

    # reformat new_grid u data to be same shape as x and y

    current_y4 = np.zeros(num_x4)
    new_x4 = []
    new_y4 = []
    new_u4 = []
    new_v4 = []
    for i in range(num_y4):
        for j in range(num_x4):
            current_y4[j] = y_line4[i]
        new_y4 = np.concatenate((new_y4, current_y4))
        new_x4 = np.concatenate((new_x4, x_line4))
        new_u4 = np.concatenate((new_u4, gridu4[i]))
        new_v4 = np.concatenate((new_v4, gridv4[i]))
        
    new_grid_centre4 = [np.mean(max_x4), np.mean(max_y4)]

    new_grid_x_range4 = [new_grid_centre4[0]-(1.4**(4-k)), new_grid_centre4[0]+(1.4**(4-k))]
    new_grid_y_range4 = [new_grid_centre4[1]-(1.4**(4-k)), new_grid_centre4[1]+(1.4**(4-k))]

    grid_x5, grid_y5 = np.mgrid[new_grid_x_range4[0]:new_grid_x_range4[1]:4j, new_grid_y_range4[0]:new_grid_y_range4[1]:4j]
    num_x5 = np.shape(grid_x5)[0]
    num_y5 = np.shape(grid_y5)[1]

    gridu5 = scipy.interpolate.griddata(points, u_mean_x, (grid_x5, grid_y5)).T
    gridv5 = scipy.interpolate.griddata(points, u_mean_y, (grid_x5, grid_y5)).T

    # now need to generate new x and y points using linspace

    x_line5 = np.linspace(new_grid_x_range4[0], new_grid_x_range4[1], num_x5)
    y_line5 = np.linspace(new_grid_y_range4[0], new_grid_y_range4[1], num_y5)

    # reformat new_grid u data to be same shape as x and y

    current_y5 = np.zeros(num_x5)
    new_x5 = []
    new_y5 = []
    new_u5 = []
    new_v5 = []
    for i in range(num_y5):
        for j in range(num_x5):
            current_y5[j] = y_line5[i]
        new_y5 = np.concatenate((new_y5, current_y5))
        new_x5 = np.concatenate((new_x5, x_line5))
        new_u5 = np.concatenate((new_u5, gridu5[i]))
        new_v5 = np.concatenate((new_v5, gridv5[i]))

    new_x = np.concatenate((new_x, new_x2, new_x3, new_x4, new_x5))
    new_y = np.concatenate((new_y, new_y2, new_y3, new_y4, new_y5))
    new_u = np.concatenate((new_u, new_u2, new_u3, new_u4, new_u5))
    new_v = np.concatenate((new_v, new_v2, new_v3, new_v4, new_v5))

    new_x = np.array(new_x)
    new_y = np.array(new_y)
    new_u = np.array(new_u)
    new_v = np.array(new_v)

    [vel_x_training, vel_y_training, u_training, v_training, dummy_t] = \
        generate_training_points(new_x, new_y, new_u, new_v, new_u)

    vel_training_points = np.hstack((vel_x_training, vel_y_training))

    training_points = np.concatenate((temp_training_points, vel_training_points))
    
    txu = []
    tyu = []
    txt = []
    tyt = []
    for i in range(0, len(vel_training_points)):
        txu.append(vel_training_points[i][0])
        tyu.append(vel_training_points[i][1])
    for i in range(0, len(temp_training_points)):
        txt.append(temp_training_points[i][0])
        tyt.append(temp_training_points[i][1])

    cylinderx = np.linspace(-0.5, 0.5, 50)
    cylindery = np.sqrt(0.5**2 - cylinderx**2)
    plt.figure(figsize=figsize)
    plt.scatter(txu, tyu, s=1.4, label="Velocity data points")
    plt.scatter(txt, tyt, s=1.4, label="Scalar data points")
    plt.plot(cylinderx, cylindery, color='k')
    plt.plot(cylinderx, -cylindery, color='k')
    axes = plt.gca()
    axes.set_aspect(1)
    axes.set_xlim([-3.0, 15.0])
    axes.set_ylim([-4.0, 4.0])
    axes.set_title("Data point distribution")
    plt.xlabel("x/c")
    plt.ylabel("y/c")
    plt.legend(loc=(1.05, 0.45))
    plt.show()
    plt.close()
    print(str(len(training_points)) + " training points in domain")

    # u, v dirichlet BCs at each point in training_points
    u_training_points = dde.PointSetBC(vel_training_points, u_training, component=0)
    v_training_points = dde.PointSetBC(vel_training_points, v_training, component=1)
    t_training_points = dde.PointSetBC(temp_training_points, t_training, component=3)
    # dirichlet BCs at walls, component indices from pde function
    bc_wall_u = dde.DirichletBC(geom, func_zeros, boundary, component=0)
    bc_wall_v = dde.DirichletBC(geom, func_zeros, boundary, component=1)
    bc_wall_t = dde.DirichletBC(geom, func_zeros, boundary, component=3)
    bc_wall_fx = dde.DirichletBC(geom, func_zeros, boundary, component=4)
    bc_wall_fy = dde.DirichletBC(geom, func_zeros, boundary, component=5)
    bc_wall_g = dde.DirichletBC(geom, func_zeros, boundary, component=6)

    data = dde.data.PDE(
        geom, pde, 
        [bc_wall_u, bc_wall_v, bc_wall_t, bc_wall_fx, bc_wall_fy, bc_wall_g, u_training_points, v_training_points, t_training_points], 
        35000, 2500, solution=None, num_test=10, 
    )

    layer_size = [2] + [100] * 7 + [7]
    activation = "tanh"
    initializer = "Glorot uniform"
    net = dde.maps.FNN(layer_size, activation, initializer)

    loss_weights = [1, 1, 1, 1, 1, 10, 10, 10, 10, 10, 10, 10, 10, 10]
    model = dde.Model(data, net)
    temp_model = deepcopy(model) # make a copy of the 'blank' model
    temp_model.compile("adam", lr=0.001, loss_weights=loss_weights) # Compile 'blank' model 

    if(k==0):
        temp_model.restore("models/prediction2-" + str(k) + "-grad-" + str(adamiters) + ".pt") # restore the previous model into the 'blank' model
        model.net.load_state_dict(temp_model.state_dict())
        model.compile("adam", lr=0.001, loss_weights=loss_weights)
    elif(k==1):
        temp_model.compile("L-BFGS-B", loss_weights=loss_weights)
        temp_model.restore("models/prediction2-" + str(k) + "-grad-" + str(adamiters + bfgsiters) + ".pt") # restore the previous model into the 'blank' model
        model.net.load_state_dict(temp_model.state_dict())
    else:
        temp_model.compile("L-BFGS-B", loss_weights=loss_weights)
        temp_model.restore("models/prediction2-" + str(k) + "-grad-" + str(bfgsiters) + ".pt") # restore the previous model into the 'blank' model
        model.net.load_state_dict(temp_model.state_dict())

    bfgsiters = int(round(((100*np.emath.logn(1.03, k+1))+3000)/1000))*1000
    model.compile("L-BFGS-B", loss_weights=loss_weights)
```
